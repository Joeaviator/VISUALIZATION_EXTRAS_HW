```python
grouped_df = df.groupby('FUELTYPE').mean(numeric_only=True)
display(grouped_df.head())
```

```python
grouped_df = grouped_df.reset_index()
display(grouped_df.head())
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='FUELTYPE', y='CO2EMISSIONS', data=grouped_df, palette='viridis')
plt.title('Average CO2 Emissions by Fuel Type')
plt.xlabel('Fuel Type')
plt.ylabel('Average CO2 Emissions')

for index, row in grouped_df.iterrows():
    plt.text(index, row.CO2EMISSIONS, round(row.CO2EMISSIONS, 2), color='black', ha="center", va="bottom")

plt.show()
```
